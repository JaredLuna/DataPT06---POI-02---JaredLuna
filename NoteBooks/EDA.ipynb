```python
#Empezamos importando las librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importamos los datos
hechos_df = pd.read_csv('../Datasets/HechosEDA.csv', encoding= 'ISO-8859-1')
victimas_df = pd.read_csv('../Datasets/VictimasEDA.csv', encoding= 'ISO-8859-1')
```

```python
#Para observaciones, vamos a grafiar las victimas y los acusados para poder encontrar algun insight
```

```python
sns.histplot(data=hechos_df['VICTIMA'])
plt.xticks(rotation= 90)
plt.ylabel(ylabel= 'Cantidad De Victimas')
plt.show()
```

```python
#Podemos observar que el top 3 de victimas son:
#1._Moto
#2._Peaton
#3._Auto
```

```python
sns.histplot(data=hechos_df['ACUSADO'])
plt.xticks(rotation= 90)
plt.ylabel(ylabel= 'Cantidad De Acusados')
plt.show()
```

```python
#Podemos observar que el top 3 de acusados son:
#1.Auto
#2._Pasajeros
#3._Cargas
```

```python
#Vamos a hacer el display de las comunas en donde hasy más accidentes
sns.barplot(data= hechos_df['COMUNA'].value_counts())
plt.xlabel(xlabel= 'Comuna')
plt.ylabel(ylabel= 'Cantidad De Accidentes')
plt.xticks(rotation= 90)
plt.show()
```

```python
#Aquí podemos observar el top de comunas que están presentes en la mayoría de accidentes, donde destacan:
#1._ La coumna 1
#2._ La comuna 4
#3._ La comuna 9
```

```python
#Ahora vamos a analizar el año que más accidentes tuvo
sns.histplot(data= hechos_df['AAAA'])
plt.xlabel(xlabel= 'Año')
plt.ylabel(ylabel= 'Cantidad de Accidentes')
```

```python
#Al parecer el top 3 esta en los primeros 3 años, pero luego despues del 2018 los incidentes bajan y luego suben en 2021
#Entonce el top 3 es
#1._ 2016
#2._ 2018
#3._ 2017
```

```python
#Ahora vamos a obbservar si existen datos nulos
nulos = hechos_df[hechos_df.isnull().any(axis= 1)]
nulos
```

```python
''' Al parecer contamos con 696 filas que contienen datos nulos, pero al parecer solo en la columna de altura. No parece que esta información 
sea tan necesaria para nuestro analisis, por lo cual no se eliminaran esas filas y solamente se les colocará una altura x en la parte del ETL'''
```

```python
''' Ahora nos centramos en el dataset de victimas, este dataset se sentra en mostrar los datos de las victimas de los accidentes'''
victimas_df
```

```python
'''Aquí podemos observar que no es un gran dataset, pero al parecer tenemos un valor en Fecha de fallecimiento que es SD, esto significa Sin Dato.
Observemos cuantos SD tenemos'''
sd_victimas = victimas_df.applymap(lambda x: 'SD' in str(x))
sd_rows = victimas_df[sd_victimas.any(axis= 1)]
sd_rows
```

```python
'''Podemos ver que contamos con 109 filas que no tienen dato, es el 15% de nuestra información,
eliminarla sería una gran perdida, por lo tanto se obtará por asignar datos a estas columnas para 
no perder la demás info.

Otra cosa importante a notar es que tenemos fechas como strings completos, por lo tanto en la EDA 
optaremos por separar esas fehcas en año, mes y día para tener dicha info más accesible'''
```

```python
'''Ahora para observar los datos, vamos a analizar en que año existieron más victimas'''
sns.histplot(data= victimas_df['AAAA'])
plt.xlabel(xlabel= 'Año')
plt.ylabel(ylabel= 'Cantidad de Victimas')
plt.show()
```

```python
'''Como era de esperar el mayor número de vicitmas también se encuentra en los años con mayor número de accidentes'''
```

```python
'''Un dato interesante es ver el sexo de la victima, primero vamos a asegurarnos que todos los valores solo sean femenino y masculino'''
victimas_df['SEXO'].unique()

''' Bueno, tenemos masculino, femenino y SD. Ahora vamos a graficarlos'''
sns.histplot(data= victimas_df['SEXO'])
plt.xlabel(xlabel= 'Sexo')
plt.ylabel(ylabel= 'Cantidad')
```

```python
'''Facilmente podemos observar que el top es el sexo masculino'''
```

```python
sns.violinplot(data= victimas_df, x= victimas_df['VICTIMA'], y= victimas_df['ROL'])

plt.xticks(rotation= 90)
plt.show()
```

```python
'''Como podemos observar de esta gráfica, para moto y auto el mayor porcentaje de victimas están cuando la victima es el conductor
Para los vehiculos de carga se mantiene casi igual si es acompañante o conductor y para movil es mayor cuando es el acompañante'''
```
